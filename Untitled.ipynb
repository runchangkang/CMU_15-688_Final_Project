```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle("advanced_data.pickle")
def changeDtypes(df):
    for i in df.columns:
        if i not in ('teamIndex','gameDate','result'):
            df[i]=df[i].astype("float64")
            
    return df
df = changeDtypes(df)
a=list(df.columns)
for i in ['teamIndex','gameDate','result']:
    a.remove(i)
```

```python
def refactorDF(df,k,a):
    newDf = pd.DataFrame(columns=df.columns)
#     print(newDf)
    for i in range(df.shape[0]-k):
        tmpDf = df.loc[i:i+k-1][a].mean()
#         print(tmpDf)
        tmpDf["gameDate"]=df.loc[i+k]["gameDate"]
        tmpDf["result"]=df.loc[i+k]["result"]
        tmpDf["teamIndex"]=df.loc[i+k]["teamIndex"]
        
        newDf.loc[newDf.shape[0]]=tmpDf
        
    return newDf
        
newdf = refactorDF(df,5,a)
```
